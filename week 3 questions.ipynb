## Heading Week 3 Quiz Questions

1. What is the primary goal of data anonymization?
   - To protect sensitive information by making it impossible or extremely difficult to identify the individuals.

2. What is the Havasupai Tribe case about?
   - Researchers from Arizona State University collected blood samples from the Havasupai Tribe to study Type 2 diabetes. However, the samples were later used without the tribe's consent for unauthorized studies.

3. What was the largest health data breach in history?
   - The 2015 Anthem Inc. breach was the largest health data breach in history.

4. What is the pulse oximeter controversy about?
   - The devices were three times more likely to miss low oxygen levels in Black patients.

5. What is the role of an Institutional Review Board (IRB) in research?
   - IRBs act as gatekeepers who review and monitor research to protect the rights, welfare, and safety of human subjects.

7. What is a conflict of interest in public health research?
   - A situation where secondary interests may unduly influence or bias research conduct or findings.

8. What is informed consent in the context of health data collection?
   - The process of providing individuals with sufficient information.

9. What is data masking in the context of de-identification of health data?
   - A technique involving replacing identifiers with random characters.

10. What is the principle of anonymity in the context of de-identification of health data?
    - A method of grouping data so that each individual record is indistinguishable.

11. What is the main concern about conflicts of interest in public health research?
    - They can introduce bias into research design and reporting.

12. What is the first step in managing conflicts of interest in public health research?
    - The first step is transparency.

13. What is the role of independent oversight in managing conflicts of interest in public health research?
    - It provides an additional layer of scrutiny by having separate communities review proposals and monitor research to ensure objectivity.

14. What is one strategy for mitigating conflicts of interest in public health research?
    - Common strategies include avoidance, divestment of financial interests, management plans, and public disclosure in publications.

15. What is the potential impact of not properly managing conflicts of interest in public health research?
    - It can lead to flawed health policies based on biased evidence, damage the reputation of the research community, and decrease public engagement.
